# Visualization of the Energy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load the data
data = np.loadtxt('energy.txt', skiprows=1)
data = pd.DataFrame(data, columns=['step', 'Ekin', 'Epot', 'Etot'])

# convert to relative energies  
data['Ekin'] -= data['Ekin'].mean()
data['Epot'] -= data['Epot'].mean()
data['Etot'] -= data['Etot'].mean()

# plot the data
plt.plot(data['step'], data['Ekin'], label='Ekin')
plt.plot(data['step'], data['Epot'], label='Epot')
plt.plot(data['step'], data['Etot'], label='Etot')
plt.xlabel('Step')
plt.ylabel('rel. Energy')
plt.legend()
plt.xlim(0, 1000)
plt.show()
```
