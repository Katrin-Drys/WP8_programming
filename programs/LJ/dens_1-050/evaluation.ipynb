# Visualization of the Energy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load the data
data = np.loadtxt('energy.txt', skiprows=1)
data = pd.DataFrame(data, columns=['step', 'time', 'temp', 'Ekin', 'Epot', 'Etot'])

# convert to relative energies  
data['Ekin'] -= data['Ekin'].mean()
data['Epot'] -= data['Epot'].mean()
data['Etot'] -= data['Etot'].mean()

# plot the data
plt.plot(data['time'], data['Ekin'], label='Ekin')
plt.plot(data['time'], data['Epot'], label='Epot')
plt.plot(data['time'], data['Etot'], label='Etot')
plt.xlabel('Time')
plt.ylabel('rel. Energy')
plt.legend()
plt.xlim(1, 40)
plt.ylim(-4000, 4000)
plt.show()
```

# Temperature

```python
plt.plot(data['time'], data['temp'])
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.xlim(1, 40)
plt.ylim(137, 143)
plt.show()
```
